```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, StratifiedKFold

import pickle

from classification import Classification
from ensemble import Ensemble
```

```python
sns.set_context('poster')
```

```python
path = "data/X_train_data.pkl" 

with open(path, 'rb') as f:
    X_train = pickle.load(f)

X_train.head(5)
```

```python
path = "data/y_train_data.pkl" 

with open(path, 'rb') as f:
    y_train = pickle.load(f)

y_train.head(5)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
y_train
```

### Train and validation split for gridsearch cross validation

```python
X_train, X_val, y_train, y_val = train_test_split(X_train,y_train[0],test_size=.25,random_state=0)
```

```python
skf = StratifiedKFold(n_splits=5,random_state=0,shuffle=True)
```

### Decision Tree

```python
params = {'min_samples_leaf':[3,5,10,15,30,50,100],
          'max_depth':[3,4,5,6,7,8,9]}
```

```python
dec_tree_1 = Classification('Decision Tree',X_train,X_val,y_train,y_val)
```

```python
dec_tree_1.get_scores(params,skf)
```

```python
params = {'min_samples_leaf':[47,48,49,50,51,52,53,54],
          'max_depth':[16,17,18,19,20,21,22,23]}
```

```python
dec_tree_2 = Classification('Decision Tree',X_train,X_val,y_train,y_val)
```

```python
dec_tree_2.get_scores(params,skf)
```
